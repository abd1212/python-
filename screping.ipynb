```python
import csv
from bs4 import BeautifulSoup
from msedge.selenium_tools import Edge, EdgeOptions
#from webdriver_manager.microsoft import EdgeChromiumDriverManager
from selenium import webdriver
#browser = webdriver.Edge(EdgeChromiumDriverManager().install())

driver = webdriver.Chrome()

def get_url(search_text,x):
    """Generate a url from search text"""
    
    
    template = 'https://www.amazon.com/s?k={}&rh=n%3A13896617011&dc&qid=1605621589&rnid=2941120011&ref=sr_nr_n_1'
    search_term = search_text.replace(' ', '+')
    
    # add term query to url
    url = template.format(search_term + '&page=' + x)
    
    # add page query placeholder
    
#     url += '&page{}'
    
        
    return url

def extract_record(item):
    """Extract and return data from a single record"""
    
#     extract description and url
    atag = item.h2.a
    description = atag.text.strip()
    url = 'https://www.amazon.com' + atag.get('href')
    
# #      extract product price
#     try:
#         price_parent = item.find('span', 'a-price')
#         price = price_parent.find('span', 'a-offscreen').text
#     except AttributeError:
#         price=''
    
#     extract rating and review count
    try:
        rating = item.i.text
#         review_count = item.find('span', {'class': 'a-size-base', 'dir': 'auto'}).text
    except AttributeError:
        rating = ''
#         review_count = ''
    print (rating)
    
#     extract review
    try:
        driver.get(url)
        soup_review = BeautifulSoup(driver.page_source, 'html.parser')
        review = soup_review.find('div',
                                  {'class': 'a-expander-content reviewText review-text-content a-expander-partial-collapse-content'}
                                 ).text.strip()
       
    except AttributeError:
        review='null'
        print(review)
    

    result = (description, rating, review, url)
    return result

def main(search_term):
    """Run main program routine"""
    
#     startup the webdriver
#     options = EdgeOptions()
#     options.use_chromium = True
#     driver = Edge(options=options)
    
    
    records = []
    url = get_url(search_term,'1')
    for page in range(1, 20):
        print(page)
        driver.get(url.format(page))
        print(url)
        soup = BeautifulSoup(driver.page_source, 'html.parser')
        results = soup.find_all('div', {'data-component-type': 's-search-result'})
        
        for item in results:
            record = extract_record(item)
            if record:
                records.append(record)
       
        print("#####",page+1)
        url = get_url(search_term,str(page+1))
    
    
    driver.close()
    # save data to csv file
    with open('review20_screp.csv', 'w', newline='', encoding='utf-8') as f:
        writer = csv.writer(f)
        writer.writerow(['Description', 'Rating', 'Review', 'Url'])
        writer.writerows(records)
        

# run program
main('computers and tablets')
```
